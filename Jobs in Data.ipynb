```python
sn.countplot(x = 'experience_level', data = jobs_in_data)
plt.xlabel('Experience Level')
plt.ylabel('Number of Employees')
plt.show()
```

```python
sn.barplot(x = 'experience_level', y = 'salary_in_usd', data = jobs_in_data)
plt.xlabel('Experience Level')
plt.ylabel('Salary in USD')
plt.show()
```

```python
sn.barplot(x = 'experience_level', y = 'salary_in_usd', data = jobs_in_data, ci = False)
plt.show()
```

```python
plt.hist(jobs_in_data['salary_in_usd'])
plt.xlabel('Salary in USD')
plt.ylabel('Number of employees')
plt.show()
```

```python
#Creating a scatter plot to compare salaries in USD and in other currencies
```

```python
#work_year, salary_in_usd
```

```python
plt.scatter(x = jobs_in_data['work_year'], y = jobs_in_data['salary_in_usd'], s = 5)
plt.xlabel('Work Year')
plt.ylabel('salary_in_usd')
plt.show()
```

```python
plt.scatter(x = jobs_in_data['salary'], y = jobs_in_data['salary_in_usd'], s = 5)
plt.xlabel('Salary in other currencies')
plt.ylabel('Salary in USD')
plt.show()
```
